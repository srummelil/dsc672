```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot
sns.set()
%matplotlib inline
```

```python
cons = pd.read_pickle('./processed/consumption_master.pkl')
cons.head().T
```

```python
cons_total = cons[['Weekdays',
                  'HolidayName',
                  'School_Day',
                  'Workday',
                  'FOOD_SVC_TOTAL',
                  'GROCERY_TOTAL',
                  'HEALTH_CARE_TOTAL',
                  'K12_TOTAL',
                  'LODGING_TOTAL',
                  'OFFICE_TOTAL',
                  'RESIDENTIAL_TOTAL',
                  'SA_RTL_TOTAL']]
cons_total.head()
```

```python
cons_total.describe().T
```

```python
cons_total.describe(exclude=[np.number]).T
```

```python
corr = cons_total.corr()
corr
```

```python
sns.pairplot(cons_total[['FOOD_SVC_TOTAL',
                  'GROCERY_TOTAL',
                  'HEALTH_CARE_TOTAL',
                  'K12_TOTAL',
                  'LODGING_TOTAL',
                  'OFFICE_TOTAL',
                  'RESIDENTIAL_TOTAL',
                  'SA_RTL_TOTAL']])
pyplot.savefig('consumption_pairplot.png', bbox_inches='tight')
```

```python
fig, ax = pyplot.subplots(figsize=(12,10))
prod_heat = sns.heatmap(ax=ax, data=corr, annot=True)
pyplot.savefig('consumption_corr_heatmap.png', bbox_inches='tight')
```
