```python
import pandas as pd
from pandas import Series
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

power_prod = pd.read_pickle("./processed/power_production.pkl")
```

```python
print(power_prod.head())
```

```python
power_prod.set_index('Date')
for source in power_prod['Source'].unique():
    print("***********************************")
    print("Source: {}".format(source))
    print(power_prod.loc[power_prod['Source'] == source].describe())
    power_source = power_prod.loc[power_prod['Source'] == source]
    plt.plot(power_source['Date'], power_source['KWH'], 'r')
    plt.title("Power Production: {}".format(source))
    plt.ylabel("KWH")
    plt.show()
```
